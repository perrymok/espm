### 1.3. eds_spim

#### 1.3.1. Fixed H

```python
roi = spim.define_ROI(xray_lines=False)
```

```python
fH_0 = spim.generate_part_fixed_H_matrix(rois=[roi])
fH_1 = spim.generate_part_fixed_H_matrix(rois=None)
fH_2 = spim.generate_part_fixed_H_matrix(rois=[])

fH_dict = {'0' : fH_0,
           '1' : fH_1,
           '2' : fH_2}

fH = spim.set_fixed_H(fH_dict)
```

```python
snmf = SmoothNMF(n_components=3,fixed_H=fH,hspy_comp=True,simplex_H=True,simplex_W=False,G=spim.model)
spim.decomposition(algorithm = snmf)
```
